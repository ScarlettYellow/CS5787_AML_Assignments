```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import os
import scipy  
from scipy import misc,ndimage
from skimage import data,filters,io,transform,feature,segmentation,restoration,util,color
from skimage.io import imread, imshow
from PIL import Image, ImageDraw
from PIL import Image
from sklearn.utils import shuffle
```

```python
img = imread('monalisa_ori.jpg')
print (img.shape)
print (img.ndim)
imshow(img)
print(img)

# gray: default 0-1
# RGB: default 0-255
img = np.array(img, dtype=np.float32)/255.0
```

```python
img
```

```python
# meshgrid
x = np.arange(0, img.shape[1])
y = np.arange(0, img.shape[0])
locations = np.meshgrid(y,x)
locations = np.stack(locations, axis=-1).reshape(-1, 2)
```

```python
locations.shape
```

```python
## sample points
num_sample_points = 10000
np.random.shuffle(locations)
sample_points = locations[:num_sample_points]
test_points = locations[num_sample_points:]
```

```python
sample_points[:10]
```

```python
## pixels match
sample_point_pixels = np.array([img[x[0], x[1]] for x in sample_points], dtype='float32')
test_point_pixels = np.array([img[x[0], x[1]] for x in test_points], dtype='float32')
```

```python
## normalize the coordinates
normalized_sample_points = sample_points.copy().astype('float32')
normalized_test_points = test_points.copy().astype('float32')

normalized_sample_points[:, 0] /= img.shape[0]
normalized_sample_points[:, 1] /= img.shape[1]
normalized_test_points[:, 0] /= img.shape[0]
normalized_test_points[:, 1] /= img.shape[1]
```

```python
test_point_pixels[0]
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=50, max_depth=8, criterion='mse')
```

```python
rfr.fit(normalized_sample_points, sample_point_pixels)
```

```python
predictions = rfr.predict(normalized_test_points)
```

```python
sample_point_pixels[:20]
```

```python
predictions[:20], predictions.shape
```

```python
## combine points and sample points

image = np.zeros_like(img)
print(image.shape)
for point, pixel in zip(sample_points, sample_point_pixels):
    image[point[0], point[1]] = pixel
for point, pixel in zip(test_points, predictions):
    image[point[0], point[1]] = pixel
```

```python
image = (image * 255).astype('uint8')
```

```python
imshow(image)
```

```python
# Sampling 
imshow(img)

samp_list = []
samp_num = 5000
rows,cols = img.shape

for i in range(samp_num):
    h = np.random.randint(0,rows) #randomly generates an int, ranges from 0 to rows
    w = np.random.randint(0,cols)
    samp = (h,w)
    samp_list.append(samp)
print ('sample amount', len(samp_list))
print ('sample list:', samp_list[:100])
```

```python
coord_list = []
coord_num = 543600
rows,cols = img.shape

for i in range(coord_num):
    h = np.random.randint(0,rows) #randomly generates an int, ranges from 0 to rows
    w = np.random.randint(0,cols)
    coord = (h,w)
    coord_list.append(coord)
print ('coord amount', len(coord_list))
print ('coord list:', coord_list[:10])
```

```python
# dataset segmentation 

#from sklearn.model_selection import train_test_split

#x_train, x_test = train_test_split(samp_list, test_size=0.3, random_state=42)
#print (len(x_train))
#print (x_train[:10])

# get y_train
y_train = []
x_train = samp_list
for i in x_train:
    label = img[i[0]][i[1]]
    y_train.append(label)
    
print ('y_train amount:', len(y_train))
print ('y_train list:', y_train[:10])

y_train=np.array(y_train)
print (y_train.shape)
print (y_train.ndim)

y_train = np.reshape(y_train, (50*100))
print (y_train.shape)
print (y_train.ndim)
print (y_train)

# get y
y = []
x = coord_list
for i in x:
    label = img[i[0]][i[1]]
    y.append(label)
    
print ('y amount:', len(y))
print ('y list:', y[:10])

y=np.array(y)
print (y.shape)
print (y.ndim)

y = np.reshape(y, (900*604))
print (y.shape)
print (y.ndim)
print (y)
```

```python
# Random Forest 
from sklearn.ensemble import RandomForestRegressor
 
rfr=RandomForestRegressor(n_estimators=5)
 
rfr.fit(x_train,y_train)   #train
 
y_pred = rfr.predict(coord_list)   #predict 

print(y_pred) 
```

```python
print(y_pred.shape)
print(y_pred.ndim)
print (type(y_pred))
#print (y_pred)

y_pred = y_pred.reshape(900, 604)

print(y_pred.shape)
print(y_pred.ndim)
print (type(y_pred))
#print (y_pred)
```
