```python
n_colors = 64
# 读取图片像素数据
img = cv2.imread("mona_gray.jpg")
#img = imread('mona_gray.jpg', as_gray=True)

#img = cv2.imread("monalisa_ori.jpg")
io.imshow(img)
```

```python
# print('tiger >>>>',tiger)
# 将图片像素数据标准化为0-1的数据并保存至数组中
img_st = np.array(img, dtype=np.float64) / 255
#print('img_st >>>>',img_st)
io.imshow(img_st)
print (img_st.ndim)

w, h, d = original_shape = tuple(img_st.shape)
print('original_shape >>>>', original_shape)
print('w,h,d >>>>', w, h, d)  # w：层数，h行数，d列数

image_array = np.reshape(img_st, (w * h, d))
# 每个点作为一个样本，维数为3，将三维数组china化为二维数组，列数不变，行数变为原行数乘层数
# print('image_array >>>>', image_array)
print('image_array shape>>>>', image_array.shape)
print (image_array)
print (len(image_array))
print (image_array.ndim)
```

```python
# Image Sampling 
#img = mean_img
plt.imshow(img)

samp_list = []
samp_num = 5000
rows,cols,dims = img.shape

for i in range(samp_num):
    h = np.random.randint(0,rows) #randomly generates an int, ranges from 0 to rows
    w = np.random.randint(0,cols)
    samp = (h,w)
    samp_list.append(samp)
print ('sample amount', len(samp_list))
print ('sample list:', samp_list[:100])
```

```python
# dataset segmentation 

#from sklearn.model_selection import train_test_split

#x_train, x_test = train_test_split(samp_list, test_size=0.3, random_state=42)
#print (len(x_train))
#print (x_train[:10])

# get y_train
y_train = []
x_train = samp_list
for i in x_train:
    label = img_st[i[0]][i[1]]
    y_train.append(label)
    
print ('y_train amount:', len(y_train))
print ('y_train list:', y_train[:10])

y_train=np.array(y_train)
print (y_train.shape)
print (y_train.ndim)

#y_train = np.reshape(y_train, (150*100))
#print (y_train.shape)
#print (y_train.ndim)
#print (y_train)
```

```python
coord_list = []
coord_num = 543600
rows,cols,dims = img.shape

for i in range(coord_num):
    h = np.random.randint(0,rows) #randomly generates an int, ranges from 0 to rows
    w = np.random.randint(0,cols)
    coord = (h,w)
    coord_list.append(coord)
print ('coord amount', len(coord_list))
print ('coord list:', coord_list[:10])
```

```python
# Random Forest 
from sklearn.ensemble import RandomForestRegressor
 
rfr=RandomForestRegressor()
 
rfr.fit(x_train,y_train)   #train
 
y_pred = rfr.predict(coord_list)   #predict 

print(y_pred) 
```
