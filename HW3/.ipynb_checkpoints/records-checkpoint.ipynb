#### Why Mean subtraction

So as to conduct **feature standardization**, namely like in the feature pre-processing of machine learning, perform mean subtraction on each dimension of the input feature vectors and then divide by the standard deviation. However, since the pixel values of each point in the natural image range from 0 to 255, the variance is roughly the same, performing the mean subtraction is enough.

Key principle: Our natural image is actually a stationary data distribution (that is, each dimension of the image follows the same distribution). Therefore, we need to subtract the statistical average of the corresponding dimensions of the data, to eliminate the common parts and to highlight the characteristics and differences among individuals. For images, this sort of normalization removes the image's average intensity, allowing us to focus more on its content.

```python
# mean subtraction 

import cv2
import numpy as np
import matplotlib.pyplot as plt

# show image
img = cv2.imread('data/mona_gray.jpg').astype(np.float32) / 255 #Standardization: pixel data into 0~1
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #show in RGB
plt.imshow(img)

# calculate the mean

mean_img = img.copy() # use img.copy() to copy img into mean_img
means = mean_img.mean() # use.mean() to acquire the mean of mean_img
print(means)

# the image matrix after mean subtraction 
mean_img = mean_img - mean_img.mean()
#print (mean_img)

# show the image after mean subtraction, the pixel brightness decreases
plt.imshow(mean_img)
```

```python
# Image Sampling 
img = mean_img
plt.imshow(img)

samp_list = []
samp_num = 5000
rows,cols,dims = img.shape

for i in range(samp_num):
    h = np.random.randint(0,rows) #randomly generates an int, ranges from 0 to rows
    w = np.random.randint(0,cols)
    samp = (h,w)
    samp_list.append(samp)
print ('sample amount', len(samp_list))
print ('sample list:', samp_list[:100])
```

```python
# dataset segmentation 

from sklearn.model_selection import train_test_split

x_train, x_test = train_test_split(samp_list, test_size=0.3, random_state=42)
print (len(x_train))
print (x_train[:10])

# get y_train
y_train = []
for i in x_train:
    label = image_gray[i[0]][i[1]]
    y_train.append(label)
    
print ('y_train amount:', len(y_train))
print ('y_train list:', y_train[:10])

# get y_test
y_test = []
for i in x_test:
    label = image_gray[i[0]][i[1]]
    y_test.append(label)
    
print ('y_test amount:', len(y_test))
print ('y_test list:', y_test[:10])
```

```python
# Standardization - Feature Scaling
# perform standardization on continuous numerical characteristics, makes its mean value = 0 and its variance = 1.

from sklearn.preprocessing import StandardScaler
st_scaler = StandardScaler()
x_train = st_scaler.fit_transform(x_train)
x_test = st_scaler.transform(x_test)

print (x_train)
```

#### Why no Normalization/Standardization 
Tree-based methods (e.g. random forest, Bagging and boosting) are not sensitive to variable size, so normalization is not required for them. Parameter-based models or distance-based models need to be normalized because parameters or distances need to be calculated.
